# POTENTIAL ENERGY CURVES (PECs)

```python
#Plotting ExoMol_ExoMol_Shift95Ma Model PECs 
oldpoten_file = 'goyutepoten.csv'
newpoten_file = 'exomol_shift95.csv'
oldpoten = pd.read_csv(f'{oldpoten_file}')
newpoten = pd.read_csv(f'{newpoten_file}')
rvals_old = oldpoten['r'].values
rvals_new = newpoten['r'].values
params = {'legend.fontsize': 30,
      'axes.labelsize': 40,
      'axes.titlesize': 40,
      'xtick.labelsize' :30,
      'ytick.labelsize': 30,
      'grid.color': 'k',
      'grid.linestyle': ':',
      'grid.linewidth': 0.5,
      'mathtext.fontset' : 'stix',
      'mathtext.rm'      : 'serif',
      'font.family'      : 'serif',
      'font.serif'       : "Times New Roman", # or "Times"          
     }
matplotlib.rcParams.update(params)
fig, ax = plt.subplots(figsize =(17, 12))
ax.plot(rvals_new,newpoten['1'], 'r-',label=r'$X^2\Pi$')
ax.plot(rvals_new,newpoten['2'], 'g-',label=r'$A^2\Sigma^+$')
ax.plot(rvals_new,newpoten['3'], 'y-',label=r'$B^2\Sigma^-$')
ax.plot(rvals_new,newpoten['4'], 'c-',label=r'$a^4\Sigma^-$')
ax.plot(rvals_new,newpoten['5'], 'm-',label=r'$b^4\Pi$')
ax.plot(rvals_new,newpoten['6'], 'b-',label=r'$D^2\Pi$')
ax.set_xlabel('Bond length ()')
ax.set_ylabel(r'Potential energy ($\rm cm^{-1}$)')
ax.set_ylim([-10000, 150000])
ax.set_xlim([0,4.5])
ax.legend(loc='best', prop={"family":"Times New Roman"})
plt.ticklabel_format(axis="y", style="sci", scilimits=(3,5))
plt.grid()
plt.savefig('exo95poten.png')
plt.show()
```
